```python
#now checking for outliers
Q1 = df['Ratings'].quantile(0.25)
Q3 = df['Ratings'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = max(0, Q1 - 1.5 * IQR)
upper_bound = min(100, Q3 + 1.5 * IQR)
((df['Ratings'] < lower_bound) | (df['Ratings'] > upper_bound)).sum()
```

```python
def convert_to_pkr(value):
    if pd.isna(value):
        return np.nan
    value = str(value).strip().lower()
    
    if 'lakh' in value:
        num = float(value.replace('lakh', '').strip())
        return num * 1e5
    elif 'crore' in value:
        num = float(value.replace('crore', '').strip())
        return num * 1e7
    elif 'million' in value:
        num = float(value.replace('million', '').strip())
        return num * 1e6
    else:
        try:
            return float(value)
        except:
            return np.nan

df['Revenue'] = df['Revenue'].apply(convert_to_pkr)
```
